```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('gambar.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

img = cv2.resize(image_rgb, (200, 200))
brightened = cv2.convertScaleAbs(img, alpha=1, beta=50)
darkened = cv2.convertScaleAbs(img, alpha=1, beta=-50)

plt.figure(figsize=(8, 4))
plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title("Image Asli")
plt.axis("off")
plt.subplot(1, 3, 2)
plt.imshow(brightened)
plt.title("Image Brightened")
plt.axis("off")
plt.subplot(1, 3, 3)
plt.imshow(darkened)
plt.title("Image Darkened")
plt.axis("off")
plt.show()

```
